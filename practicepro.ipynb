```python
n=int(input("Enter a number:"))
for i in range(1,n+1):
    for j in range(1,n+1):
        if i==(n//2)+1:
            print('*',end=' ')
        elif i==1 and j not in(1,n):
            print('*',end=' ')
        elif j==1 and i!=1:
            print('*',end=' ')
        elif j==n and j!=1:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```
